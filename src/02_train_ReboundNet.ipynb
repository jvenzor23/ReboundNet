```python
def create_x(data, nfeatures):
    # creating a numpy tensor to hold the play data
    np_mat = np.array(data.drop(['Unnamed: 0','GAME_ID', 'GAME_EVENT_ID', 'play_id', 'player_id1','player_id2',
                                    'teammate_flag', 't_after_shot',"group"],
                               axis = 1))
    
    # setting parameters
    ngames = np.max(np.array(data['play_num']))
    nPlayers = 10
    nOpponents = 5
    
    # initializing the tensor
    np_tensor = np.zeros((ngames, nPlayers, nOpponents, nfeatures))
    
    for row in range(len(np_mat)):
        np_tensor[int(np_mat[row][0]) - 1][int(np_mat[row][1]) - 1][(int(np_mat[row][2]) - 1)%5] = np_mat[row][3:]
    
    np_tensor = np_tensor.astype('float32')
    
    return(np_tensor)
```

```python
def create_y(data):
    targets = np.array(data["target"])
    
    y = np.zeros((targets.shape[0], 12))
    for idx, target in enumerate(list(targets)):
        y[idx][target - 1] = 1
    
    return(y)
```

```python
import tensorflow.keras.backend as K
def playerteamopt(y_true, y_pred):
    y_true_team = K.batch_flatten(K.sum(y_true[:,0:5], axis = 1))
    y_pred_team = K.batch_flatten(K.sum(y_pred[:,0:5], axis = 1))
#     return 0.5*tf.keras.backend.categorical_crossentropy(y_true, y_pred) + \
#            0.5*tf.keras.backend.binary_crossentropy(y_true_team, y_pred_team)
    
#     return 0.5*(1 - K.mean(tf.linalg.diag_part(K.dot(y_pred, K.transpose(tf.dtypes.cast(y_true, tf.float32)))))) + \
#            0.5*tf.keras.backend.binary_crossentropy(y_true_team, y_pred_team)
    
    return 0.5*(K.mean(K.pow(1 - tf.linalg.diag_part(K.dot(y_pred, K.transpose(tf.dtypes.cast(y_true, tf.float32)))), 1.5))) + \
            0.5*tf.keras.backend.binary_crossentropy(y_true_team, y_pred_team)
```

```python
def better_crossentropy(y_true, y_pred):
    return 1 - K.mean(tf.linalg.diag_part(K.dot(y_pred, K.transpose(tf.dtypes.cast(y_true, tf.float32)))))
```

```python
def team_crossentropy(y_true, y_pred):
    y_true_team = K.batch_flatten(K.sum(y_true[:,0:5], axis = 1))
    y_pred_team = K.batch_flatten(K.sum(y_pred[:,0:5], axis = 1))
    return tf.keras.backend.binary_crossentropy(y_true_team, y_pred_team)
```

```python
def def_probs(y_true, y_pred):
    y_pred_team = K.sum(y_pred[:,0:5], axis = 1)
    return K.mean(y_pred_team)
```

```python
def def_freq(y_true, y_pred):
    y_true_team = K.sum(y_true[:,0:5], axis = 1)
    return K.mean(y_true_team)
```

```python
def top_k_categorical_accuracy2(y_true, y_pred, k=2):
    return K.mean(K.in_top_k(y_pred, K.argmax(y_true, axis=-1), k), axis=-1)

def top_k_categorical_accuracy3(y_true, y_pred, k=3):
    return K.mean(K.in_top_k(y_pred, K.argmax(y_true, axis=-1), k), axis=-1)
```

# Feature Importance (Permutation)

```python
import random
```

```python
nfeatures = X_train2.shape[-1]
nSamples = X_train2.shape[0]
X_train_feature_imp = np.zeros(X_train2.shape)
accuracy = []
for iCnt in range(nfeatures):
    X_train_feature_imp = X_train2.copy()
    for jCnt in range(nSamples):
        jumbled_vals = X_train2[:,:,:,iCnt][jCnt].ravel()
        jumbled_vals2 = random.sample(list(jumbled_vals), len(jumbled_vals))
        jumbled_vals3 = np.array(jumbled_vals2).reshape(X_train2[:,:,:,iCnt][jCnt].shape)
        X_train_feature_imp[:,:,:,iCnt][jCnt] = jumbled_vals3
    
    # training the model
    model = tf.keras.Sequential()

    model.add(layers.Conv2D(100, 
                            kernel_size=(1,1), 
                            strides=(1,1), 
                            input_shape = (10,5,13)))
    model.add(layers.Activation('relu'))
    model.add(layers.Dropout(0.2))
    model.add(layers.Conv2D(50, 
                            kernel_size=(1,1), 
                            strides=(1,1)))
    model.add(layers.Activation('relu'))
    model.add(layers.Conv1D(100, 
                             kernel_size=1, 
                             strides=1))
    model.add(layers.Activation('relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.2))

    import tensorflow.keras.backend as K

    def channelPool(x):
        return .5*K.max(x,axis=2) + .5*K.mean(x,axis=2)

    model.add(layers.Lambda(channelPool))

    model.add(layers.BatchNormalization())
    model.add(layers.Conv1D(100, 
                            kernel_size=1, 
                            strides=1))
    model.add(layers.Activation('relu'))
    model.add(layers.Dropout(0.2))
    model.add(layers.Conv1D(50, 
                            kernel_size=1, 
                            strides=1))
    model.add(layers.Activation('relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv1D(1, 
                             kernel_size=1, 
                             strides=1))
    model.add(layers.Activation('relu'))

    model.add(layers.Flatten())
    model.add(layers.Activation('softmax'))
    
    model.compile(loss=['categorical_crossentropy'], metrics=['categorical_accuracy', top_k_categorical_accuracy2,
                                                    top_k_categorical_accuracy3], optimizer='adam')
    
    history = model.fit(x = X_train_feature_imp, y = y_train2,
                   validation_data = (X_validation2, y_validation2),
                   epochs = 5,
                   # callbacks=callbacks, # Early stopping
                   verbose = 1)
    
    accuracy += [history.history['categorical_accuracy'][-1]]
```
